```python
def plot_class_distribution_pie(class_name):  
  class_counts = df[class_name].value_counts()
  total_instances = df.shape[0]

  class_percentages = [(count / total_instances) * 100 for count in class_counts]

  plt.figure(figsize=(6, 8))
  plt.pie(class_percentages, labels=df[class_name].unique(), autopct='%1.1f%%', startangle=45,  shadow=True)
  plt.axis('equal')  
  plt.title(class_name + ' Distribution')
  plt.show()
  #plt.savefig(class_name + '.png')
```

```python
plot_class_distribution_pie('Species')
```

```python
plot_class_distribution_pie('Genus')
```

```python
plot_class_distribution_pie('Family')
```

```python
cont_df = df.iloc[: , :-4]
plt.figure(figsize=(20,20))
for i in range(1, 23):
    plt.subplot(6, 4, i)
    sns.histplot(cont_df[cont_df.columns[i-1]], bins=14, kde_kws={'bw_method': 0.1}, color="r")
```

```python
fig = plt.figure(figsize=(30,20))
for i,feature in enumerate(df.columns[:-4]):
  grouped_data = df.groupby('Species')[feature].apply(list)

  class_labels = grouped_data.index.tolist()
  attribute_values = grouped_data.values.tolist()

  i+=1
  plt.subplot(6, 4, i)
  plt.boxplot(attribute_values, labels=class_labels)
  plt.xlabel('Species')
  plt.xticks(rotation=45, fontsize=8)
  plt.ylabel(feature)
  plt.tight_layout()
plt.show()
```

```python
corr = df.corr()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))

# Create a custom divergin palette
cmap = sns.diverging_palette(250, 15, s=75, l=40,
                            n=9, center="light", as_cmap=True)

plt.figure(figsize=(16, 12))
sns.heatmap(corr, mask=mask, center=0, annot=True,
            fmt='.1f', square=True, cmap=cmap)

plt.show();
```
